```python
df.columns = df.columns.str.replace(r'\n', ' ', regex=True)  # Remove newline characters
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
df.columns = df.columns.str.replace("  ", " ")  # Replace double spaces with single space

print(df.columns)  # Verify cleaned column names
```

```python
df["States/ Union Territories"].unique()
```

```python
df = df.drop_duplicates()
```

```python
df["States/ Union Territories"] = df["States/ Union Territories"].str.strip().str.title()
```

```python
df["Total"] = df[["Intermediate/ Senior Secondary Schools", "High/ Secondary Schools", 
                  "Upper Primary Schools", "Primary Schools", "Pre-Primary Schools"]].sum(axis=1)
```

```python
df.fillna(0, inplace=True) 
```

```python
df.to_csv("cleaned_noseInst.csv", index=False)
```

```python
Q1 = df.iloc[:, 1:].quantile(0.25)
Q3 = df.iloc[:, 1:].quantile(0.75)
IQR = Q3 - Q1
```

```python
outlier_condition = (df.iloc[:, 1:] < (Q1 - 1.5 * IQR)) | (df.iloc[:, 1:] > (Q3 + 1.5 * IQR))
```

```python

df_cleaned = df[~outlier_condition.any(axis=1)]
```

```python
df_cleaned.to_csv("Cleaned_noseInst.csv", index=False)
```

```python
plt.xlabel("States")
plt.ylabel("Total Schools")
plt.title("Top 10 States with the Most Schools")
plt.xticks(rotation=45)
plt.show()
```

```python
school_distribution = df[["Intermediate/ Senior Secondary Schools", "High/ Secondary Schools", 
                          "Upper Primary Schools", "Primary Schools", "Pre-Primary Schools"]].sum()

# Plot Pie Chart
plt.figure(figsize=(8,8))
plt.pie(school_distribution, labels=school_distribution.index, autopct="%1.1f%%", colors=["blue", "green", "red", "purple", "orange"])
plt.title("School Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Primary Schools"], y=df["High/ Secondary Schools"], hue=df["States/ Union Territories"], palette="viridis")
plt.xlabel("Primary Schools")
plt.ylabel("High/Secondary Schools")
plt.title("Primary Schools vs. High Schools")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.iloc[:, 1:].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap Between School Types")
plt.show()
```

```python
plt.figure(figsize=(12,6))
for col in ["Primary Schools", "Upper Primary Schools", "High/ Secondary Schools", "Intermediate/ Senior Secondary Schools"]:
    sns.lineplot(x=df["States/ Union Territories"], y=df[col], label=col)
plt.xticks(rotation=90)
plt.xlabel("States")
plt.ylabel("Number of Schools")
plt.title("School Growth Trend Across States")
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.iloc[:, 1:], palette="Set2")
plt.xticks(rotation=45)
plt.title("Distribution of Schools by Category")
plt.xlabel("School Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["Total"], bins=10, kde=True, color="teal")
plt.xlabel("Total Schools")
plt.ylabel("Number of States")
plt.title("Distribution of Schools Across States")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x="States/ Union Territories", y="Total", data=df, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("School Count Density per State")
plt.show()
```
